## **COR DO SOLO**

---


As cores do solo so usadas para adicionar beleza e simbolismo  pintura corporal, pinturas rupestres e outras formas de arte. As cores derivadas do solo variam de amarelo e laranja a vermelho e marrom. Cada cor tem seu prprio significado e significado em diferentes culturas. Eles foram usados pela primeira vez pelo povo aborgine da Austrlia, que usava ocre, um pigmento amarelo-avermelhado, para pintar seus corpos para cerimnias e cerimnias. Escavaes em cavernas na Austrlia indicam que o ocre foi amplamente usado pelos primeiros australianos entre 57.000 e 71.000 anos atrs. Os nativos americanos Chusmas da Califrnia tambm usaram subtons em suas obras de arte. Eles usam hematita, um pigmento vermelho, para pintar listras e manchas vermelhas em seus corpos e tocas. Alm da hematita, a tribo Chusmas tambm utilizava a argila branca para pintura corporal e rituais de cura. As cores do solo tm sido usadas h sculos e ainda so usadas hoje. (GUIMARES, 2016).


Ainda hoje, as cores da terra so proeminentes no artesanato, onde os pigmentos da terra so usados como corantes. O projeto Cores da Terra (CARVALHO et al., 2007) foi desenvolvido pelo Departamento de Solos da Universidade Federal de Viosa (UFV) para ensinar a fazer tintas de solo para pintar casas. As marcas de roupas havaianas so conhecidas por tingir todas as suas roupas com argila vermelha. Minerais so comumente usados na indstria farmacutica como ingredientes ou como excipientes para exibir propriedades teraputicas. Tambm  muito utilizado em spas e centros de beleza (CARRETERO & POZO, 2009). Alm das propriedades fsicas, mecnicas e trmicas dos minerais, propriedades visuais como pigmentos e opacificantes so amplamente utilizadas para melhorar as propriedades organolpticas de produtos farmacuticos (CARRETERO & POZO, 2009).

Cientistas russos comearam a prestar ateno  cor do solo no incio do sculo XIX e comearam a descrev-la em termos de matiz, luminosidade e saturao na dcada de 1920 (ZAKHAROV, 1927). Nos Estados Unidos, as propriedades de cor do solo no foram includas no sistema de classificao at 1914. Segundo Guimares (2016), quando a Diviso de Solos do Departamento de Agricultura dos Estados Unidos (USDA) publicou uma lista de 22 nomes de fundos de cores de solo, no havia padres de cores at ento. O USDA, juntamente com a Munsell Company, preparou uma srie de tabelas de cores para uso no Soil Survey Manual de 1951 (SOIL SURVEY STAFF, 1993). Desde ento, o sistema de cores Munsell (MUNSELL SOIL COLOR COMPANY, 1950) e uma nova lista de nomes tm sido usados e recomendados pelo USDA para determinar a cor do solo (SIMONSON, 1993). O sistema comea a ser adotado por pesquisadores de solo em todo o mundo.

Em 1949, a MUNSELL SOIL COLOR COMPANY criou o Sistema Americano de Classificao de Solos. Em 1950, a EMBRAPA comeou a desenvolver um novo sistema de classificao de solos brasileiro que evoluiu desse antigo sistema de classificao. (EMBRAPA, 2013) A primeira verso do SiBCS foi lanada em 1999 e desde ento foi revisada vrias vezes. O diagrama de Munsell (MUNSELL SOIL COLOR COMPANY, 1950) serve como padro para classificar as cores do solo no Sistema Brasileiro de Classificao de Solos - SiBCS.

```python
# extrai da imagem 'solo1.jpg' o cluster k=1 do k-means e mostra rgb
import numpy as np # importa numpy
import cv2 #import opencv
from matplotlib import pyplot as plt # importa pyplot
from sklearn.cluster import KMeans # importa k-means

img = cv2.imread('solo1.jpg') # leia a imagem
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #converte para rgb
plt.imshow(img)
plt.show()

Z = img.reshape((-1,3)) # remodela para uma lista de pixels
Z = np.float32(Z) # converter para np.float32
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) # define critrios, nmero de clusters(K) e aplica kmeans()
K = 1 # nmero de clusters
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS) # converte para valores de 8 bits
center = np.uint8(center) # converte para uint8

res = center[label.flatten()] # converte de volta para a imagem de 3 canais da imagem de 1 canal
 
res2 = res.reshape((img.shape)) # mostra a imagem
plt.imshow(res2)
plt.show() 

# converter rgb para munsell
import colorsys # importar colorsys
def rgb_to_munsell(r, g, b): # define funo
    h, l, s = colorsys.rgb_to_hls(r/255.0, g/255.0, b/255.0) #converter rgb para hls
    h = h*360 # converter h de 0-1 para 0-360
    if h < 20: # se h for menor que 20
        hue = "R" # matiz  vermelho
    elif h < 40: # se h for menor que 40
        hue = "R-Y" # matiz  vermelho-amarelo
    elif h < 75: # se h for menor que 75
        hue = "Y" # matiz  amarelo
    elif h < 155: # se h for menor que 155
        hue = "G-Y" # matiz  verde-amarelo
    elif h < 190: # se h for menor que 190
        hue = "G" # matiz  verde
    elif h < 260: # se h for menor que 260
        hue = "B-G" # matiz  verde-azulado
    elif h < 290: # se h for menor que 290
        hue = "B" # matiz  azul
    elif h < 335: # se h for menor que 335
        hue = "P-B" # matiz  roxo-azul
    else:
        hue = "P" # matiz  roxo
    if l < 0.2: # se l for menor que 0,2
        value = "10" # valor  10
    elif l < 0.4: # se l for menor que 0.4
        value = "20" # valor  20
    elif l < 0.6: # se l for menor que 0.6
        value = "30"  # valor  30
    elif l < 0.8: #  se l for menor que0.8
        value = "40" # valor  40
    else: 
        value = "50" # valor  50
    if s < 0.1: # se s for menor que 0,1
        chroma = "0" # croma  0
    elif s < 0.2: # se s for menor que 0.2
        chroma = "1" # croma  1
    elif s < 0.3: # se s for menor que 0.3
        chroma = "2" # croma  2
    elif s < 0.4: # se s for menor que 0.4
        chroma = "3" # croma  3
    elif s < 0.5: # se s for menor que 0.5
        chroma = "4" # croma  4
    elif s < 0.6: # se s for menor que 0.6
        chroma = "5" # croma  5
    elif s < 0.7:  # se s for menor que 0.7
        chroma = "6" # croma  6
    elif s < 0.8: # se s for menor que 0.8
        chroma = "7" # croma  7
    elif s < 0.9:  # se s for menor que 0.9
        chroma = "8" # croma  8
    elif s < 1.0: # se s for menor que 1.0
        chroma = "9" # croma  9
    return hue + " " + value + " " + chroma # retorna valor de matiz e croma

# salva a tabela rgb do cluster k=1 k-means
import csv # importar csv
with open('solo1.csv', 'w', newline='') as file: # abrir arquivo
    writer = csv.writer(file) # escreve no arquivo
    writer.writerow(["R","G","B"]) # escreve no arquivo
    writer.writerow(center[0]) # escreve no arquivo

    # imprime a tabela rgb do cluster k=1 k-means
    print("R,G,B") # imprime no console
    print(center[0]) # imprime no console

    
#salve a tabela munsell do cluster k=1 k-means
import csv # importar csv
with open('solo1munsell.csv', 'w', newline='') as file: # open file
    writer = csv.writer(file) # escreve no arquivo
    writer.writerow(["Munsell"]) # escreve no arquivo
    writer.writerow([rgb_to_munsell(center[0][0],center[0][1],center[0][2])]) # escreve no arquivo
    # imprime a tabela munsell do cluster k=1 k-means
    print("Munsell") # imprime no console
    print(rgb_to_munsell(center[0][0],center[0][1],center[0][2])) # imprime no console

```
